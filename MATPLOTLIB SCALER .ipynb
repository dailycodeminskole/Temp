```python
# x_val = [-3,-2,-1,0,1,2,3]
# y_val = [9,4,1,0,1,4,9]
# plt.plot(x_val,y_val)

# using numpy to get the same points
x_val = np.arange(-3,3,0.1)
y_val = x_val**2
plt.plot(x_val,y_val)
```

```python
# More precise ticks
x_val = np.arange(-3,3,0.1)
y_val = x_val**2
plt.plot(x_val,y_val)
plt.xticks(np.arange(-3,3,0.3),rotation=90) #more partitions in x-axis, rotation for clear visibility
plt.show()
```

```python
# addding labels and title
```

```python
x_val = np.arange(-3,3,0.1)
y_val = x_val**2
plt.plot(x_val,y_val)

plt.title("Square")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks(np.arange(-3,3,0.3),rotation=90) #more partitions in x-axis, rotation for clear visibility
plt.show()
```

```python
# changing theme color,
x_val = np.arange(-3,3,0.1)
y_val = x_val**2
plt.plot(x_val,y_val,'ro') #  * + --  other can also be used


plt.title("Square")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.xticks(np.arange(-3,3,0.3),rotation=90) #more partitions in x-axis, rotation for clear visibility
plt.show()

```

```python
#  zooming IN  --- accessing only limited part of the plot
x_val = np.arange(-3,3,0.1)
y_val = x_val**2
plt.plot(x_val,y_val)


plt.xlim(1,3)#limited view
plt.title("Square")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
# zooming out plot -- Birds Eyeview
x_val = np.arange(-3,3,0.1)
y_val = x_val**2
plt.plot(x_val,y_val)


plt.xlim(-10,10)#limited view
plt.ylim(0,5)#limited view for y axis zoom in 

plt.title("Square")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
# Multile plots in same canvas ( default in differnt colors)
x_val = np.arange(-3,3,0.1)
y_1 = x_val**2
y_2 = x_val**3

plt.plot(x_val,y_1)
plt.plot(x_val,y_2)
```
