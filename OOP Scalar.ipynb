# understanding the class variable/attribute

```python
 class Student:
        def __init__(self,name):
            self.name = name
            counter = 0
            counter = counter+1
            self.roll = counter
        def __str__(self):
            return "The name of Student is : {} and roll no {}".format(self.name , self.roll)
```

```python
s1 = Student('Rahul')
print(s1)
```

```python
s1 = Student('Ramesh')
print(s1)
```

```python
# the counter will reset to 0 each time the object is created
# but will not be changed if the attibute belongs to the class
```

```python
 class Student:
        counter = 0 #class variable
        def __init__(self,name):
            self.name = name            
            Student.counter = Student.counter+1 #sharing a class variable
            self.roll = Student.counter
        def __str__(self):
            return "The name of Student is : {} and roll no {}".format(self.name , self.roll)
```

```python
s1 = Student('Rahul')
print(s1)
s2 = Student('Ramesh')
print(s2)
```

```python
print(s1.counter) #check in local scope , but is not there , so it check in the class , and we have given the access 
print(s2.counter)
```

```python
print(Student.counter)
```

```python
print(s1.roll)# this is the value we set
print(s2.roll)
```

```python
 class Student:
        counter = 100 #class variable
        def __init__(self,name):
            self.name = name            
            Student.counter = Student.counter+1 #sharing a class variable
            self.roll = Student.counter
        def __str__(self):
            return "The name of Student is : {} and roll no {}".format(self.name , self.roll)
        
        def intro(self):
            print("My name is {}".format(self.name))
s1 = Student('Rahul')
s1.intro()
```
